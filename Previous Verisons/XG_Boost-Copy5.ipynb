```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

Now that everything is an integer or float it should be easier to model. 

```python
data_test = pd.read_csv('/Users/aakashtandel/Desktop/test_weather3e.csv', index_col='Date')
data_test.head()
```

```python
data_test.info()
```

```python
data_test = data_test.rename(columns={'year': 'Year'})
data = data.drop(['year', 'NumMosquitos', 'Latitude', 'Longitude'], axis=1)
data_test = data_test.drop(['Latitude', 'Longitude'], axis=1)
```

```python
data = data.reindex_axis(sorted(data.columns), axis=1)
data_test = data_test.reindex_axis(sorted(data_test.columns), axis=1)
```

```python
print (data.info())
print (data_test.info())
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import log_loss, accuracy_score
from collections import Counter
```

```python
X = data.drop(['WnvPresent'], axis=1)
y = data[['WnvPresent']]
```

```python
X.head()
```

```python
y.head()
```

```python
sample = pd.DataFrame()
sample['Id'] = data_test['Id']
sample.head()
```

```python
data_test = data_test.drop('Id', 1)
```

### Simple Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X, y)
dt_pred = dt.predict(data_test)
dt_prob = dt.predict_proba(data_test)[:,1]
dt_nodes = dt.tree_.node_count
print ('Decision Tree')
print ('Number of Nodes: ', str(dt_nodes))
sample['WnvPresent_DecisionTree'] = dt_prob
```

```python
sample.head()
```

### AdaBoost Classifier 

```python
adaboost = AdaBoostClassifier(
            base_estimator=DecisionTreeClassifier(max_depth=1), 
            algorithm='SAMME', n_estimators=1000, random_state=100)
adaboost.fit(X, y)
ada_pred = adaboost.predict(data_test)
ada_pred_proba = adaboost.predict_proba(data_test)[:,1]
print ('AdaBoost')
sample['WnvPresent_Ada'] = ada_pred_proba
sample.head()
```

### Gradient Boost Classifier

```python
gbc = GradientBoostingClassifier(
            max_depth=1, warm_start=True, 
            n_estimators=1000, random_state=100)
gbc.fit(X, y)
gbc_pred = gbc.predict(data_test)
gbc_pred_proba = gbc.predict_proba(data_test)[:,1]
print ('Gradient Boost')
sample['WnvPresent_gradientboost'] = gbc_pred_proba
sample.head()
```

### XGBoost 

```python
import xgboost as xgb
```

```python
params = {'objective':'binary:logistic', 'max_depth':2, 'silent':1, 'eta':1}
num_rounds = 5
dtrain = xgb.DMatrix(X, label=y)
dtest = xgb.DMatrix(data_test)
```

```python
boosted = xgb.train(params, dtrain, num_rounds)
```
