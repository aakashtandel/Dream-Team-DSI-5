```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import log_loss, accuracy_score
from collections import Counter
```

```python
X = data.drop(['WnvPresent'], axis=1)
y = data[['WnvPresent']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = StandardScaler().fit_transform(X.values)
scaled_features_df = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)
scaled_features_df.head()
```

```python
sample = pd.DataFrame()
sample['Id'] = data_test['Id']
sample.head()
```

```python
data_test = data_test.drop('Id', 1)
scaled_test = StandardScaler().fit_transform(data_test.values)
scaled_test_df = pd.DataFrame(scaled_test, index=data_test.index, columns=data_test.columns)
scaled_test_df.head()
```

### Simple Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(scaled_features_df, y)
dt_prob = dt.predict_proba(scaled_test)[:,1]
dt_nodes = dt.tree_.node_count
print ('Decision Tree')
print ('Number of Nodes: ', str(dt_nodes))
sample['WnvPresent_DecisionTree'] = dt_prob
```

```python
sample.head()
```

### AdaBoost Classifier 

```python
adaboost = AdaBoostClassifier(
            base_estimator=DecisionTreeClassifier(max_depth=1), 
            algorithm='SAMME', n_estimators=1000, random_state=100)
adaboost.fit(scaled_features_df, y)
ada_pred = adaboost.predict(scaled_test)
ada_pred_proba = adaboost.predict_proba(scaled_test)[:,1]
print ('AdaBoost')
sample['WnvPresent_Ada'] = ada_pred_proba
sample.head()
```

### Gradient Boost Classifier

```python
gbc = GradientBoostingClassifier(
            max_depth=1, warm_start=True, 
            n_estimators=1000, random_state=100)
gbc.fit(scaled_features_df, y)
gbc_pred = gbc.predict(scaled_test)
gbc_pred_proba = gbc.predict_proba(scaled_test)[:,1]
print ('Gradient Boost')
sample['WnvPresent_gradientboost'] = gbc_pred_proba
sample.head()
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc
```

```python
print (scaled_features_df.shape)
print (y.shape)
y.head()    
```

```python
gbm = xgb.XGBClassifier()
parameters = {'n_jobs':[-1,4], 
              'objective':['binary:logistic'],
              'learning_rate': [0.05, 0.1, 0.3, 1], 
              'max_depth': [2, 6, 9, 12],
              'min_child_weight': [11],
              'silent': [1],
              'subsample': [0.8, 0.9, 1.0],
              'colsample_bytree': [0.7, 0.9, 1.0],
              'n_estimators': [5,10,100,100], 
              }
clf = GridSearchCV(gbm, parameters, n_jobs=5, 
                   cv=StratifiedKFold(y['WnvPresent'], n_folds=5, shuffle=True), verbose=2,
                   scoring='roc_auc', refit=True)
clf.fit(scaled_features_df, y)
# print (clf.best_params_)
# print (clf.best_estimator_)
```

```python
best = clf.best_estimator_
preds = best.predict_proba(data_test)[:,1]
print (preds)
```

```python
sample['WnvPresent_xgboost_cv'] = preds
sample.head()
```
