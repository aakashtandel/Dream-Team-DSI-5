```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
data = pd.read_csv('/Users/aakashtandel/Desktop/train_weather.csv', index_col=0)
data.head()
```

```python
data.info()
```

```python
def T_stripper(string):
    return string[1:4]
data['Trapp'] = data['Trap'].apply(T_stripper)
data.Trapp = data.Trapp.astype(int)
data.head() 
```

```python
data.info()
```

Now that everything is an integer or float it should be easier to model. 

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import log_loss, accuracy_score
from collections import Counter
```

```python
X = data.drop(['Trap', 'WnvPresent'], axis=1)
y = data[['WnvPresent']]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
print ('Train Label Distribution')
print ( Counter(y_train) )
print ('Test Label Distribution')
print ( Counter(y_test) )
```

### Simple Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_prob = dt.predict_proba(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
dt_log_loss = log_loss(y_test, dt_pred)
dt_nodes = dt.tree_.node_count
print ('Decision Tree')
print ('Accuracy: ', str(dt_accuracy))
print ('Log Loss: ', str(dt_log_loss))
print ('Number of Nodes: ', str(dt_nodes))
```

### AdaBoost Classifier 

```python
adaboost = AdaBoostClassifier(
            base_estimator=DecisionTreeClassifier(max_depth=1), 
            algorithm='SAMME', n_estimators=1000, random_state=100)
adaboost.fit(X_train, y_train)
ada_pred = adaboost.predict(X_test)
ada_pred_proba = adaboost.predict_proba(X_test)
ada_accuracy = accuracy_score(y_test, ada_pred)
ada_log_loss = log_loss(y_test, ada_pred)
print ('AdaBoost')
print ('Accuracy: ', str(ada_accuracy))
print ('Log Loss: ', str(ada_log_loss))
```

### Gradient Boost Classifier

```python
gbc = GradientBoostingClassifier(
            max_depth=1, warm_start=True, 
            n_estimators=1000, random_state=100)
gbc.fit(X_train, y_train)
gbc_pred = gbc.predict(X_test)
gbc_pred_proba = gbc.predict_proba(X_test)
gbc_accuracy = accuracy_score(y_test, gbc_pred)
gbc_log_loss = log_loss(y_test, gbc_pred)
print ('Gradient Boost')
print ('Accuracy: ', str(gbc_accuracy))
print ('Log Loss: ', str(gbc_log_loss))
```
