```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
train = pd.read_csv('../project_4_local/csvs/train.csv')
weather = pd.read_csv('../project_4_local/csvs/weather.csv')
test = pd.read_csv('../project_4_local/csvs/test.csv')
spray = pd.read_csv('../project_4_local/csvs/spray.csv')
sample = pd.read_csv('../project_4_local/csvs/sampleSubmission.csv')
train.columns = [col.lower() for col in train.columns]
test.columns = [col.lower() for col in test.columns]



# Make column names lowercase.
weather.columns = [i.lower() for i in weather.columns.values]
# Turn date column to datetime format.
#weather.date = pd.to_datetime(weather.date)
# Replace 'T', which means trace amounts, with 0.
weather.preciptotal = weather.preciptotal.str.strip()
 # very small amounts of rain can impact mosquito hatches
weather.replace(to_replace=['T'], value=[0.005], inplace=True)
# Drop meaningless columns. Codesum may be useful, but we'll start without it.
weather.drop(['snowfall', 'depth', 'water1', 'sunrise', 'sunset', 'codesum'], axis=1, inplace=True)
```

```python
weather = weather.replace('M', np.NaN)
weather = weather.replace('-', np.NaN)
weather.tmax = weather.tmax.fillna(method='ffill')
weather.tmin = weather.tmin.fillna(method='ffill')
weather.depart = weather.depart.fillna(method='ffill')
weather.heat = weather.heat.fillna(method='ffill')
weather.cool = weather.cool.fillna(method='ffill')
weather.preciptotal = weather.preciptotal.fillna(method='ffill')
weather.tavg = weather.preciptotal.fillna(method='ffill')
weather.wetbulb = weather.wetbulb.fillna(method='ffill')
weather.stnpressure = weather.stnpressure.fillna(method='ffill')
weather.sealevel = weather.sealevel.fillna(method='ffill')
weather.avgspeed = weather.avgspeed.fillna(method='ffill')
```

```python
weather.isnull().sum()
```

```python
weather = weather.astype({'tavg':np.float, 'depart':np.float, 'wetbulb':np.float, 'heat':np.float, 'cool':np.float,
                     'preciptotal':np.float, 'stnpressure':np.float, 'sealevel':np.float, 'avgspeed':np.float},
                    inplace=True)
```

```python
## First Method.
species_dict = {'CULEX ERRATICUS':1, 'CULEX PIPIENS':2, 'CULEX PIPIENS/RESTUANS':3, 'CULEX RESTUANS':4,
                'CULEX SALINARIUS':5, 'CULEX TARSALIS':6, 'CULEX TERRITANS':7, 'UNSPECIFIED CULEX':4}
species_dict2 = {'CULEX ERRATICUS':1, 'CULEX PIPIENS':2, 'CULEX PIPIENS/RESTUANS':3, 'CULEX RESTUANS':4,
                'CULEX SALINARIUS':5, 'CULEX TARSALIS':6, 'CULEX TERRITANS':7, 'UNSPECIFIED CULEX':1}
```

```python
test.species = test.species.map(species_dict2)
```

```python
from sklearn import ensemble, preprocessing
lbls = preprocessing.LabelEncoder()
lbls.fit(list(train['trap'].values) + list(test['trap'].values))
train['trap'] = lbls.transform(train['trap'].values)
test['trap'] = lbls.transform(test['trap'].values)

lbls.fit(list(train['species'].values) + list(test['species'].values))
train['species'] = lbls.transform(train['species'].values)
test['species'] = lbls.transform(test['species'].values)

X = train.drop(['trap', 'species', 'address', 'block', 'street', 'addressnumberandstreet', 'latitude', 'longitude',
               'addressaccuracy', 'nummosquitos', 'wnvpresent', 'heat_x', 'heat_y', 'stnpressure_x', 'stnpressure_y'], axis=1)
y = train['wnvpresent']
```

```python
clf = ensemble.RandomForestClassifier(n_jobs=-1, n_estimators=1000, min_samples_split=2, random_state=5)
clf.fit(X, y)
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(clf, X, y, cv=3)
```
